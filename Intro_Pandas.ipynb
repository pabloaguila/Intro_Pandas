## Filtrar y Ordenar
Dos operaciones que se suelen necesitar con frecuencia en el análisis de datos.

```python
#imaginemos que queremos analizar los datos de una única aerolínea. Digamos AA. 
df_AA = df.loc[df["carrier"]=="AA"]
df_AA.info()
```

```python
#También podemos ordenar el dataset, 
#por ejemplo que vaya de menor a mayor retraso a la salida.
df_AA.sort_values("dep_delay", )
```

Vemos que al final del dataset hay vuelos cuyo retraso es "NaN" que significa "Not a Number". Esto ocurre cuando no se encuentra el valor de una columna par un vuelo en particular. Podemos deshacernos de estos vuelos usando el comando .dropna(). Este comando tira todos los vuelos que tengan alguna columna sin datos.

```python
df_AA.sort_values("dep_delay", ).dropna()
```

```python
#También podemos calcular el retraso de salida promedio de los vuelos de AA
df_AA["dep_delay"].mean()
```

## Cambio de unidades
A lo largo de esta notebook estuvimos viendo dataset cuyas filas correspondían a vuelos. En este caso decimos que las unidades de los datasets son vuelos. Ahora si queremos estudiar otro tipo de grupo de unidades, por ejemplo aerolíneas, podemos usar el comando .group_by().

```python
df_aerolineas = df.groupby("carrier")
```

Ahora df_aerolineas hace operaciones agrupando por aerolínea. ¿Que quiere decir esto? Por ejemplo, si calculamos el valor medio del retraso a la salida obtenemos el promedio para cada una de las aerolíneas

```python
df_aerolineas["dep_delay"].mean()
```

## Análisis de múltiples variables

Imaginemos que queremos analizar la relación entre dos variables. Por ejemplo, podemos intuir que debe haber una correlación entre los retrasos a la llegada y a la salida de los vuelos. Para ver esto podemos hacer un gráfico de dispersión.

```python
import matplotlib.pyplot as plt #esta librería nos permite hacer gráficos facheros.
#primero graficamos los datos de los vuelos
df.plot(kind="scatter", x="dep_delay", y="arr_delay", grid=True, label="Vuelos") 
#agregamos una recta que muestra donde son iguales ambos retrasos
plt.plot([-100,1400], [-100,1400], "r",)
#definimos los límites del gráfico y las etiquetas de los dejes
plt.xlim(-100,1400)
plt.ylim(-100,1400)
plt.xlabel("Retraso a la salida (min)")
plt.ylabel("Retraso a la llegada (min)")
```

Es interesante ver que hay algunos vuelos que a pesar de haber salido con retraso llegan sin retraso a su destino. Podemos calcular una nueva cantidad que sea el tiempo ganado en el aire.



```python
df["tiempo_ganado"] = df["dep_delay"]-df["arr_delay"]
```

```python
#Nuevamente podemos ordenar los datos en base a una columna
df.dropna().sort_values("tiempo_ganado")
```

```python
#Podemos hacer un histograma que nos muestra la cantidad de vuelos con una determinada cantidad de tiempo ganado 
df["tiempo_ganado"].plot.hist(grid=True)
plt.xlabel("Tiempo ganado en el aire (min)")
plt.ylabel("Cantidad de vuelos")
```
